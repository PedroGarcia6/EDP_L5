### Questão 1

Considerando que a temperatura de um vértice em um grid depende da temperatura de seus 4 vizinhos, tem-se

$$u_t(x,y)=c\cdot(u(x,y-\Delta y)+u(x,y+\Delta y)+u(x-\Delta x, y)+u(x+\Delta x, y)-4u(x,y)).$$

Via aproximação de Taylor,

$$u(x+\Delta x,y)\approx u+\Delta x\cdot u_x+\frac{\Delta x^2}{2}\cdot u_{xx},$$
$$u(x-\Delta x,y)\approx u-\Delta x\cdot u_x+\frac{\Delta x^2}{2}\cdot u_{xx},$$
$$u(x,y+\Delta y)\approx u+\Delta y\cdot u_y+\frac{\Delta y^2}{2}\cdot u_{yy},$$
$$u(x,y-\Delta y)\approx u-\Delta y\cdot u_y+\frac{\Delta y^2}{2}\cdot u_{yy}.$$

Somando tudo,

$$u(x+\Delta x,y)+u(x-\Delta x,y)+u(x,y+\Delta y)+u(x,y-\Delta y)\approx4u+\Delta x^2u_{xx}+\Delta y^2u_{yy}.$$

Logo,

$$u_t(x,y)=c\cdot(\Delta x^2u_{xx}+\Delta y^2u_{yy}).$$

Considerando $\Delta x=\Delta y$ e que
$$0<\lim_{\Delta x\to0}c\cdot\Delta x^2<\infty,$$

isto é, tende a uma constante $\tilde{c}$ finita, conclui-se que

$$u_t(x,y)=c\cdot\Delta x^2(u_{xx}+u_{yy})\implies\lim_{\Delta x\to0}c\cdot\Delta x^2(u_{xx}+u_{yy})=\tilde{c}\cdot\Delta u. \blacksquare$$

### Questão 2

Nos itens abaixo faz-se uso das seguintes propriedades:

Propriedade 1: Se $f(x)$ é uma função par em $[-L,L]$, então $f(x)\sin\left(\frac{k\pi x}{L}\right)$ é uma função ímpar e tem-se

$$b_n=0\implies f(x)=\frac{a_0}{2}+\sum_{n=1}^{\infty}a_n\cos\left(\frac{n\pi x}{L}\right).$$

Propriedade 2: Se $f(x)$ é uma função ímpar em $[-L,L]$, então $f(x)\cos\left(\frac{k\pi x}{L}\right)$ também é uma função ímpar e tem-se

$$a_n=0\implies f(x)=\sum_{n=1}^{\infty}b_n\sin\left(\frac{n\pi x}{L}\right).$$

a) $a:[−\pi,\pi],a(x)=\sin(5x)$.

Trata-se de uma função ímpar. Então, vale a Propriedade 2:

$$a(x)=\sum_{n=1}^{\infty}b_n\sin\left(\frac{n\pi x}{L}\right),$$

com $L=\pi$ e

$$b_n=\frac{1}{\pi}\int_{-\pi}^{\pi}\sin(5x)\sin(nx)dx=-\frac{10\sin(n\pi)}{(n^2-25)\pi}.$$

Assim,

$$a(x)=\sum_{n=1}^{\infty}-\frac{10\sin(n\pi)}{(n^2-25)\pi}\sin(nx).$$

Porém, nota-se que $-\frac{10\sin(n\pi)}{(n^2-25)\pi}=0,\forall\ n\neq5$. Logo, nosso interesse está no termo $n=5$, quando

$$b_{5}=\frac{1}{\pi}\int_{-\pi}^{\pi}\sin{(5x)}\sin{(5x)}dx=\frac{1}{\pi}\int_{-\pi}^{\pi}\sin²{(5x)}dx=1.$$

Portanto,

$$a(x)=\sum_{n=1}^{\infty}1\cdot\sin(5x)=\sin(5x).$$

b) $b:[−\pi,\pi],b(x)=\sin(5x+\alpha)$

Calculam-se os coeficientes de Fourier:

$$a_n=\frac{1}{\pi}\int_{-\pi}^{\pi}\sin(5x+\alpha)\cos(nx)dx=-\frac{2n\sin(n\pi)sin(\alpha)}{(n²-25)\pi},$$
$$b_n=\frac{1}{\pi}\int_{-\pi}^{\pi}\sin(5x+\alpha)\sin(nx)dx=-\frac{10\cos(\alpha)\sin(n\pi)}{(n²-25)\pi}$$

e, consequentemente, $a_0=0$. Porém, nota-se que $a_n=b_n=0,\forall\ n\neq5$. Logo, nosso interesse está no termo $n=5$, quando

$$a_n=\frac{1}{\pi}\int_{-\pi}^{\pi}\sin(5x+\alpha)\cos(5x)dx=\sin(\alpha),$$
$$b_n=\frac{1}{\pi}\int_{-\pi}^{\pi}\sin(5x+\alpha)\sin(5x)dx=\cos(\alpha).$$

Portanto,

$$b(x)=\sum_{n=1}^{\infty}\sin(\alpha)\cos(5x)+\cos(\alpha)\sin(5x)=\sin(5x+\alpha)$$

c) $c:[−\pi,\pi],c(x)=\left\{\begin{matrix}
0, \text{ se } x\leq0 \\
1, \text{ se } x>0\end{matrix}\right.$

Calculam-se os coeficientes de Fourier:
$$a_0=\frac{1}{\pi}\left[\int_{-\pi}^{0}0\ dx+\int_{0}^{\pi}1\ dx\right]=\frac{1}{\pi}\int_{0}^{\pi}1\ dx=1,$$
$$a_n=\frac{1}{\pi}\left[\int_{-\pi}^{0}0\cdot\cos(nx)dx+\int_{0}^{\pi}1\cdot\cos(nx)dx\right]=\frac{1}{\pi}\int_{0}^{\pi}\cos(nx)dx=\frac{\sin(\pi n)}{\pi n},$$
$$b_n=\frac{1}{\pi}\left[\int_{-\pi}^{0}0\cdot\sin(nx)dx+\int_{0}^{\pi}1\cdot\sin(nx)dx\right]=\frac{1}{\pi}\int_{0}^{\pi}\sin(nx)dx=\frac{1-\cos(\pi n)}{\pi n}.$$

Porém, nota-se que que $a_n=0,\forall\ n>0$. Logo,
$$c(x)=\frac{1}{2}+\sum_{n=1}^{\infty}\frac{1-\cos(\pi n)}{\pi n}\sin(nx).$$

d) $d:[−1,1],d(x)=x$

Trata-se de uma função ímpar. Então, vale a Propriedade 2:

$$d(x)=\sum_{n=1}^{\infty}b_n\sin\left(\frac{n\pi x}{L}\right),$$

com $L=1$ e

$$b_n=\int_{-1}^{1}x\sin(n\pi x)dx=-\frac{2n\pi\cos(n\pi)}{n^2 \pi^2}=-\frac{2(-1)^n}{n\pi}.$$

Assim,

$$d(x)=\sum_{n=1}^{\infty}-\frac{2(-1)^n}{n\pi}\sin(n\pi x).$$

$$u_t(x,t)=u(x-1,t)-2u(x,t)+u(x+1,t).$$

Equivalentemente,

$$u_t(x,t)=(l\circledast u)(x,t)\implies\widehat{u_t}(k,t)=\widehat{l}(k)\cdot\widehat{u}(k,t)\implies\widehat{u}(k,t)=e^{\widehat{l}(k)t}\widehat{u_0}(k).$$

Portanto,

Aplica-se Fourier na equação dada:

$$u_t=-cu_x\implies\widehat{u_t}(k,t)=\widehat{-cu_x}(k,t)=-cik\widehat{u}(k,t)\implies\widehat{u}(k,t)=e^{-cikt}\widehat{u}(k,0)=e^{-cikt}\widehat{f}(k).$$

Assim,

$$u(x,t)=\sum_{k\in\mathbb{Z}}\widehat{u}(k,t)e^{ikx}=\sum_{k\in\mathbb{Z}}\widehat{f}(k)e^{ik(x-ct)}.$$

Mas

$$f(x-ct)=\sum_{k\in\mathbb{Z}}\widehat{f}(k)e^{ik(x-ct)},$$

isto é, $u(x,t)=f(x-ct).$

### Questão 2

Eis abaixo um programa que retorna a frequência dominante em hertz de um arquivo de áudio digital.

```python
from numpy import argmax
from numpy.fft import fft
from scipy.io import wavfile

def afinador(audio):
    taxa, dado = wavfile.read(audio)
    dado_transf = fft(dado)
    maximo = argmax(abs(dado_transf))
    freq_dom = taxa*maximo/len(dado)
    return freq_dom

print('Frequência dominante: {}Hz'.format(afinador('CordaViolao2.wav')))
```

A frequência de 330Hz corresponde à nota musical Mi na 4ª oitava (E4).
